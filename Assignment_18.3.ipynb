```python
#Question: You take the SAT and score 1100. The mean score for the SAT is 1026 and the standard
#deviation is 209. How well did you score on the test compared to the average test taker?

#Formula, Z = (x-mu)/sigma

x = 1100
mu = 1026
sigma = 209

z = (x-mu)/sigma
z
print('It means the score was',z, 'std devs above the mean')
```
